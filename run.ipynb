```python
from __future__ import print_function, absolute_import
from datetime import datetime, timedelta
from gm.api import *
import pandas as pd

# 设置API token
set_token("59e9c8cd18899c191fa133d4cdb08d66efba3dd9")

# 获取今天的日期
today = datetime.today()
# 计算60天前的日期
sixty_days_ago = today - timedelta(days=60)

# 将日期格式化为字符串，用于API请求
today_str = today.strftime("%Y-%m-%d")
sixty_days_ago_str = sixty_days_ago.strftime("%Y-%m-%d")

# 获取沪深300指数的成分股
index_code = 'SHSE.000300'  # 沪深300指数的代码
constituents = stk_get_index_constituents(index_code)

# 获取成分股的代码列表
list1 = list(constituents['symbol'])

# 获取最近60天的历史数据
history_data = history(symbol=list1, frequency='1d', start_time=sixty_days_ago_str, end_time=today_str, df=True)

# 定义一个函数来计算60日最低价并判断是否处于新低
def find_60_day_low_stocks(history_data):
    low_stocks = []
    for symbol in history_data['symbol'].unique():
        stock_data = history_data[history_data['symbol'] == symbol]
        # 计算60日最低价
        stock_data.loc[:, '60_day_low'] = stock_data['low'].rolling(window=60).min()
        # 获取最新一天的数据
        latest_data = stock_data.iloc[-1]
        # 判断是否处于60日新低
        if latest_data['low'] == latest_data['60_day_low']:
            low_stocks.append((symbol, latest_data['eob'], latest_data['low']))
    return low_stocks

# 找出最近60日新低的股票
low_stocks = find_60_day_low_stocks(history_data)

# 将结果转换为DataFrame
low_stocks_df = pd.DataFrame(low_stocks, columns=['股票代码', '日期', '最低价'])

# 打印结果
print(low_stocks_df)
```
